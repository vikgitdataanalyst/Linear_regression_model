```python
# Importing libraries
import numpy as np
from sklearn import model_selection

# Loading data
data=np.loadtxt('data.csv', delimiter=",")
data

```

```python
x=data[:,0].reshape(-1,1) # Reshaping coverting into 1D array
y=data[:,1]
```

```python
# Splitting into training and testing
x_train,x_test,y_train,y_test=model_selection.train_test_split(x,y)

from sklearn.linear_model import LinearRegression

# Fitting data
alg1=LinearRegression()
alg1.fit(x_train,y_train)
```

```python
# Finding 'm' and 'c'
m=alg1.coef_[0]
c=alg1.intercept_

# Plotting graph
import matplotlib.pyplot as plt
x_line=np.arange(23,70,0.1)
y_line=m*x_line+c
plt.scatter(x_train,y_train,c="b")
plt.plot(x_line,y_line,c="g")
plt.show()
```

#### In figure most of data points are closer to the actual line y=m*x+c

## Testing data

```python
m=alg1.coef_[0]
c=alg1.intercept_
# print(m)
# print(c)

import matplotlib.pyplot as plt
x_line=np.arange(23,70,0.1)
y_line=m*x_line+c
plt.scatter(x_test,y_test,c="b")
plt.plot(x_line,y_line,c="g")
plt.show()
```

#### Testing data points are somehow closer to the actual line
#### Showing results are not to bad

```python
train_score=alg1.score(x_train,y_train)
test_score=alg1.score(x_test,y_test)
train_score,test_score

# Score is depends on data we have, it lies between 0 and 1, known as coefficient of determination
# Sometimes we happy with 0.6 and sometime we will not satisfied with 0.9
```
